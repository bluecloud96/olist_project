## 많이 팔린 물건 카테고리

```python
read_category_top10_sql = """
WITH category_english as
(SELECT
    PC.product_category_name_english AS english_name,
    P.product_category_name AS name,
    P.product_id
FROM
    category_name_translation PC
LEFT JOIN
    products P
ON
    PC.product_category_name = P.product_category_name)
SELECT
    category_english.english_name AS category,
    count(order_items.order_id) AS COUNT
FROM
    order_items
LEFT JOIN
    category_english
ON
    order_items.product_id = category_english.product_id
GROUP BY
    category
ORDER BY
    COUNT DESC,
    category
LIMIT 10;
"""
```

```python
category_top10 = pd.read_sql(read_category_top10_sql,engine)
```

```python
category_top10 = category_top10.set_index("category")
```

```python
bars = category_top10.plot(kind="bar",color ="green")
plt.title("판매량 top10 카테고리")
plt.xlabel("카테고리")
plt.ylabel("판매량")

for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval,yval, ha = "center", va = "bottom")
```

# 가설 검증

```python
read_review_avg_sql = """
with review_item as
(SELECT
	reviews.review_id,
    reviews.order_id,
    reviews.review_score,
    order_items.product_id
FROM
	reviews
LEFT JOIN
	order_items
ON
	reviews.order_id = order_items.order_id),
    review_product AS
(SELECT
    review_item.review_score,
    products.product_description_lenght,
    products.product_photos_qty
FROM
	review_item
LEFT JOIN
	products
ON
	review_item.product_id = products.product_id)
SELECT
	review_product.review_score,
    AVG(review_product.product_description_lenght) AS AVREAGE_description_lenght,
    AVG(review_product.product_photos_qty) AS AVREAGE_photos
FROM
	review_product
GROUP BY
	review_product.review_score
ORDER BY
	review_product.review_score;
"""
```

```python
review_avg = pd.read_sql(read_review_avg_sql,engine)
```

```python
review_avg
```

```python
review_avg = review_avg.set_index("review_score")
review_avg_description = review_avg[["AVREAGE_description_lenght"]]
review_avg_photos = review_avg[["AVREAGE_photos"]]
```

```python
review_avg_description.plot(kind = "bar")
plt.title('리뷰 점수별 평균 물건 설명문 길이')
plt.xlabel("리뷰 점수")
plt.ylabel("평균 물건 설명문 길이")
_=plt.xticks(rotation = 0)
```

```python
review_avg_photos.plot(kind="bar")
plt.title('리뷰 점수별 평균 물건 사진 수')
plt.xlabel("리뷰 점수")
plt.ylabel("평균 물건 사진 수")
plt.xticks(np.arange(1, 6, 1))
_=plt.xticks(rotation = 0)
```

```python
read_review_avg_group_sql='''WITH review_item AS (
    SELECT
        reviews.review_id,
        reviews.order_id,
        reviews.review_score,
        order_items.product_id
    FROM
        reviews
    LEFT JOIN
        order_items
    ON
        reviews.order_id = order_items.order_id
),
review_product AS (
    SELECT
        review_item.review_score,
        products.product_description_lenght,
        products.product_photos_qty
    FROM
        review_item
    LEFT JOIN
        products
    ON
        review_item.product_id = products.product_id
),
review_grouped AS (
    SELECT
        CASE
            WHEN review_score >= 3 THEN '3점 이상'
            ELSE '3점 이하'
        END AS score_group,
        product_description_lenght,
        product_photos_qty
    FROM
        review_product
)
SELECT
    score_group,
    AVG(product_description_lenght) AS AVERAGE_description_lenght,
    AVG(product_photos_qty) AS AVERAGE_photos
FROM
    review_grouped
GROUP BY
    score_group
ORDER BY
    score_group;'''
```

# t-test를 해보자

```python
read_3_over_sql = """
WITH review_item AS (
    SELECT
        reviews.review_id,
        reviews.order_id,
        reviews.review_score,
        order_items.product_id
    FROM
        reviews
    LEFT JOIN
        order_items
    ON
        reviews.order_id = order_items.order_id
),
review_product AS (
    SELECT
        review_item.review_score,
        products.product_description_lenght,
        products.product_photos_qty
    FROM
        review_item
    LEFT JOIN
        products
    ON
        review_item.product_id = products.product_id
)
    SELECT
        review_score,
        product_description_lenght,
        product_photos_qty
    FROM
        review_product
    WHERE
        review_score >= 3 AND
        product_description_lenght is not null
"""
```

```python
read_3_less_sql = """
WITH review_item AS (
    SELECT
        reviews.review_id,
        reviews.order_id,
        reviews.review_score,
        order_items.product_id
    FROM
        reviews
    LEFT JOIN
        order_items
    ON
        reviews.order_id = order_items.order_id
),
review_product AS (
    SELECT
        review_item.review_score,
        products.product_description_lenght,
        products.product_photos_qty
    FROM
        review_item
    LEFT JOIN
        products
    ON
        review_item.product_id = products.product_id
)
    SELECT
        review_score,
        product_description_lenght,
        product_photos_qty
    FROM
        review_product
    WHERE
        review_score < 3 AND
        product_description_lenght is not null

"""
```

```python
score_3_over = pd.read_sql(read_3_over_sql,engine)
```

```python
score_3_less = pd.read_sql(read_3_less_sql, engine)
```

# t-test시행

```python
## product_photos_qty
```

```python
from scipy import stats
_, p_value_levene = stats.levene(score_3_over['product_photos_qty'], score_3_less['product_photos_qty'])
if p_value_levene > 0.05:
    print(f" p-value: {p_value_levene}, 등분산 가정 만족")
else:
    print(f" p-value: {p_value_levene}, 이분산 가정 만족")
```

```python
t_statistic, p_value = stats.ttest_ind(
    a=score_3_over['product_photos_qty'],
    b=score_3_less['product_photos_qty'],
    alternative="two-sided",
    equal_var=False
)
```

```python
print(f"p-value: {p_value}")
print(f"귀무 가설 기각: {p_value < 0.05}")
```
